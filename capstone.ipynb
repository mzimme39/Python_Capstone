```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python

df = pd.read_csv(r"C:\Users\morit\OneDrive\Desktop\Uni\Master\Semester 2\Capstone Project\Data\out.csv", sep='\\')

with open('output.txt', 'w') as f:
    f.write(str(df))

print(df.head())
```

```python
# Überprüfen der Größe des DataFrames
print('DataFrame shape:', df.shape)

# Überprüfen der Struktur des DataFrames
print('DataFrame info:')
print(df.info())
```

```python
# Überprüfen der Verteilung der Zielvariablen
print('Verteilung der Zielvariablen:')
print(df['Gruende'].value_counts())
```

```python
# Überprüfen auf leere Werte
print('Leere Werte pro Spalte:')
print(df.isnull().sum())
```

```python
# Berechnen der Textlänge pro Zeile
text_length = df['Gruende'].str.len()

# Visualisierung der Verteilung der Textlänge
plt.hist(text_length, bins=50)
plt.xlabel('Textlänge')
plt.ylabel('Anzahl der Texte')
plt.show()
```

```python
# Extrahieren Sie die Schlüsselwörter, die in Ihren Daten vorkommen
keywords = ' '.join(df['Gruende']).lower()

# Erstellen Sie eine WordCloud mit den extrahierten Schlüsselwörtern
wordcloud = WordCloud(width=800, height=800, background_color='white').generate(keywords)

# Visualisieren Sie die WordCloud
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```
